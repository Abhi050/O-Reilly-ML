```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("housing.csv")
print(df.head())
print(df.describe())

df.hist(bins=50, figsize=(20,15))
plt.show()
```

```
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(df,test_size=0.2,random_state=42)
```

```
df.plot(kind="scatter",x="longitude",y="latitude")
plt.show()
```

```
df.plot(kind="scatter",x="longitude",y="latitude",alpha=0.1)
plt.show()
```

```
df.plot(kind="scatter",x="longitude",y="latitude",alpha=0.1,s=df["population"]/100, label="population", figsize=(10,7),c="median_house_value",cmap=plt.get_cmap("jet"),colorbar=True)
plt.legend()
plt.show()
```

```
corr_matrix=df.corr()
#PEarson's r
print(corr_matrix)

#to increase correaltion among features
df["rooms_per_household"]=df["total_rooms"]/df["households"]
df["bedroom_per_household"]=df["total_bedrooms"]/df["households"]
df["population_per_household"]=df["population"]/df["households"]

corr_matrix2=df.corr()
print(corr_matrix2)
#correlation incease
```
